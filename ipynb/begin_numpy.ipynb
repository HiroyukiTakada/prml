5 minutes to numpy
==========================

```python
# Python's list is useless for numerical computation!
a = [0.1,  0.2,  0.3]
b = [1.0,  2.0,  3.0]
c = a + b
print(c)
```

```python
# Let's use numpy!
import numpy as np

a = np.array([0.1, 0.2, 0.3])
b = np.array([1.0, 2.0, 3.0])
c = a + b
print(c)
```

```python
# The array indices begins 0.
print( c[0], c[1], c[2] )
print( c.shape )
print( c.size )
```

```python
# Inner product
print("a =", a)
print("b =", b)
z = a[0]*b[0] + a[1]*b[1] + a[2]*b[2]
print( z )

z = a.dot(b)
print("Dot product of a and b is", z)
```

```python
d = a
print(d)

# d is not a copy but a 'view' of a
d[0] = 99
print("d = ", d)
print("a = ", a)
```

```python
# 2d array as a matrix?
ma = np.array( [[20.03,  20.13,  20.23],  [21.03,  21.13,  21.23]] )
print(ma)

print("The shape of ma is", ma.shape[0], "by", ma.shape[1])

x = ma[0,2]
print("ma[0,2] is", x)
```

```python
# Matrix transposition (view)
mb = ma.T
print(mb)

print("The shape of mb is", mb.shape[0], "by", mb.shape[1])
mb[2,0] = 9999
print(ma)
```

```python
# Extract a row as an array
q = ma[1, :]  # the 2nd row
print(q)
```

```python
# Extract a column as an array
p = ma[:, 0]    # the 1st column
print(p)
```

```python
# Extract columns as a 2d array
r = ma[:, [0,2]]  # the 1st and 2nd columns
print(r)
```

```python
# Identity matrix
me = np.eye(3)
print(me)

# Zero matrix
siz = (3,5)  # tuple
m0 = np.zeros(siz)
print(m0)

# Ones matrix
m1 = np.ones((3,2))
print(m1)
```

```python
# elementwise multiplications
mh = m1 * 0.5
print(mh)
print(mh * mh)
```
